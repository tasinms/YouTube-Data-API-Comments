```python
import pandas as pd

df = pd.read_csv('WikipediaMovieDatasetFinalNew.csv')

df.head()
```

```python
# Correlation between Budget and Box Office
correlation = df['Budget (float)'].corr(df['Box Office (float)'])
correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting a Histogram of the 'Running Time (int)' column
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Running Time (int)', bins=10)
plt.title('Distribution of Running Time')
plt.xlabel('Running Time')
plt.ylabel('Count')
plt.show()

# Calculating the Mean and Standard Deviation of the 'Budget (float)' column
budget_mean = df['Budget (float)'].mean()
budget_std = df['Budget (float)'].std()
print(f"Mean Budget: {budget_mean}")
print(f"Standard Deviation of Budget: {budget_std}")
```
